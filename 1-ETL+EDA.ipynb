Análisis visuales de las tecnologías por provincia

```python


# Selecciona las columnas de interés
columnas_interes = ['ADSL', 'Cablemódem', 'Dial Up', 'Fibra óptica', '4G', '3G', 'Telefonía Fija', 'Wireless', 'Satelital']

# Agrupa por 'Provincia' y realiza el recuento de "SI" y "NO" en las columnas seleccionadas
conteo_por_provincia = df_pen_internet_100hogares.groupby('Provincia')[columnas_interes].apply(lambda x: x.eq("SI").sum()).reset_index()

# Cambia los nombres de las columnas para que sean más descriptivos
conteo_por_provincia.columns = ['Provincia'] + [col for col in columnas_interes]

# Imprime el resultado
conteo_por_provincia
```

```python

# Selecciona las columnas de interés
columnas_interes = ['ADSL', 'Cablemódem', 'Dial Up', 'Fibra óptica', '4G', '3G', 'Telefonía Fija', 'Wireless', 'Satelital']

# Agrupa por 'Provincia' y realiza el recuento de "SI" en las columnas seleccionadas
conteo_por_provincia = df_pen_internet_100hogares.groupby('Provincia')[columnas_interes].apply(lambda x: x.eq("SI").sum()).reset_index()

# Configura el gráfico de barras agrupadas
plt.figure(figsize=(12, 6))
bar_width = 0.15
index = np.arange(len(conteo_por_provincia))
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'gray', 'orange']

for i, col in enumerate(columnas_interes):
    plt.bar(index + i * bar_width, conteo_por_provincia[col], bar_width, label=col, color=colors[i])

# Personaliza el gráfico
plt.xlabel('Provincia')
plt.ylabel('Cantidad de "SI"')
plt.title('Cantidad de tecnologías por provincia(Cada 100 hogares)')
plt.xticks(index + bar_width * (len(columnas_interes) / 2), conteo_por_provincia['Provincia'], rotation=90)
plt.legend()
plt.tight_layout()

# Muestra el gráfico
plt.show()
```

```python

# Selecciona las columnas de interés
columnas_interes = ['ADSL', 'Cablemódem', 'Dial Up', 'Fibra óptica', '4G', '3G', 'Telefonía Fija', 'Wireless', 'Satelital']

# Agrupa por 'Provincia' y realiza el recuento de "SI" en las columnas seleccionadas
conteo_por_provincia = df_pen_internet_100hogares.groupby('Provincia')[columnas_interes].apply(lambda x: x.eq("NO").sum()).reset_index()

# Configura el gráfico de barras agrupadas
plt.figure(figsize=(12, 6))
bar_width = 0.15
index = np.arange(len(conteo_por_provincia))
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'gray', 'orange']

for i, col in enumerate(columnas_interes):
    plt.bar(index + i * bar_width, conteo_por_provincia[col], bar_width, label=col, color=colors[i])

# Personaliza el gráfico
plt.xlabel('Provincia')
plt.ylabel('Cantidad de "SI"')
plt.title('Cantidad de hogares sin las tecnologías por provincia(Cada 100 hogares)')
plt.xticks(index + bar_width * (len(columnas_interes) / 2), conteo_por_provincia['Provincia'], rotation=90)
plt.legend()
plt.tight_layout()

# Muestra el gráfico
plt.show()
```

|Comparación de técnologías más "Actuales" con las "desactualizadas"

```python
# Aquí crearemos datos de ejemplo para representar la idea.
provincias = df_pen_internet_100hogares['Provincia']
accesos_por_hogares = df_pen_internet_100hogares['Accesos por cada 100 hogares']

# Definimos el ancho de las barras
bar_width = 0.35

# Coordenadas x para las barras
x = np.arange(len(provincias))

# Crear las barras para accesos por cada 100 hogares
plt.bar(x, accesos_por_hogares, width=bar_width, label='Accesos por cada 100 hogares', alpha=0.7)

# Configuración de la visualización
plt.xlabel('Provincia')
plt.ylabel('Accesos por cada 100 hogares')
plt.title('Comparación de Accesos por cada 100 hogares por Provincia')
plt.xticks(x, provincias, rotation=45)  # Rotamos las etiquetas para una mejor visualización
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
